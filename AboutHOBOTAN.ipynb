### 1. HOBOについて
HOBOとは、「High Order Binary Optimization」の略で、量子コンピューティングで組合せ最適化問題を扱う際に利用するQUBO（Quadratic Unconstrained Binary Optimization）を拡張した手法です。<br>
名前の通りQUBOは、問題を２つの量子ビットの相互作用までを許容した「２次の定式化」で表現します。<br>
一方で、HOBOは3次以上の定式化を行うことができます。<br>

QUBOで最適化問題を解く場合、n個の量子ビットのうち、1つの量子ビットを「１」、それ以外を「０」にするOne-Hotエンコーディングを利用して定式化を行います。<br>
この方法では、最適化問題の制約条件を表現する際に多くの量子ビットが必要になってしまいます。<br>
HOBOでは、3次以上の定式化を利用できるため、One-Hotエンコーディングではなく、整数変数を利用して制約条件を表現できるため、必要な量子ビット数を削減できます。<br>

### 2. HOBOソルバーについて
全く知らなかったのですが、ゲート型量子コンピュータでは、QAOA（Quantum Approximate Optimization Algorithm）という最適化アルゴリズムを用いてHOBOを扱うことができます。<br>
その際、複数の量子ビットの相互作用を扱うため、$RZ^{\otimes n}$ ゲートの実装が必要になります。<br>
このようなゲートは複数の2量子ビット間の相互作用の組み合わせで実現できます。<br>

#### 3-1. 日本地図の彩色問題をHOBOで考える
彩色問題の条件は以下として定式化を行います。<br>
条件：<br>
- 各県は1色で塗ること
- 隣接する県は異なる色で塗ること
- 沖縄県と鹿児島県/北海道と青森県は異なる色で塗ること
<br>

1つのエリアに割り当てる色を2つの変数（4色）で表し、隣接する県が同じ色の場合にコストが上がるように定式化を行います。<br>

```python
import networkx as nx
import numpy as np

n = 47 # 頂点の数

# 各都道府県の隣接情報を取得
edge_list = []
for i in range(1,48):
  for j in jp.adjacent(i):
    edge_list.append((i, j))

edge_list.append((1, 2)) #北海道と青森は隣接とする。
edge_list.append((46, 47)) #沖縄と鹿児島は隣接とする。

print(edge_list)

# グラフを定義
options = {'node_color': '#efefef','node_size': 20,'with_labels':'True'}
G = nx.Graph()
G.add_nodes_from(nx.path_graph(n))
G.add_edges_from(edge_list)
nx.draw(G, **options)
```

取得した隣接情報を元に<br>
A県に割り当てられる色を2量子ビットで $q_i, q_{i+1}$ 、B県を $q_j, q_{j+1}$ として<br>
<br>
$((q_i - q_j)^2-1)^2 \times ((q_{i+1} - q_{j+1})^2-1)^2$<br>
<br>
のように定式化します。
<br>
<br>
量子ビットを用意して、上記の定式化していくので、量子アニーリングのSDK (Tytan)、HOBOTANをインストールします。<br>

```python
!pip -q install tytan
!pip install -U git+https://github.com/ShoyaYasuda/hobotan
```

```python
from hobotan import *

A = np.triu(nx.to_numpy_array(G)) # 隣接する県を２重に取得しないように、上三角行列 = 隣接行列を取得する。

# 量子ビットの用意
# 各県ごとに4色を示す２量子ビット用意する。(都道府県コード)
q = symbols_list([47, 2], 'q{}_{}')

# 式の設定 (HOBO)
# 隣接する県の間で同じ色を示すビットが選択されるとコストが高くなるようにする。
H = 0
for i in range(n):
  for j in range(n):
    if(A[i,j] == 1):
      H += ((q[i][0] - q[j][0]) **2 - 1) **2 * ((q[i][1] - q[j][1]) **2 - 1) **2

# 色が少ない方が良いをペナルティとして設定。ex) red :00 = 0, blue: 01 = 1として、redが選択されやすくする。
k = 0.1
for i in range(n):
  H += k * (q[i][0] + 2*q[i][1])

# コンパイル・サンプリング
hobo, offset = Compile(H).get_hobo()
print(f'offset\n{offset}')

solver = sampler.MIKASAmpler(seed=0)
result = solver.run(hobo, shots=100)

# アニーリング結果表示
print("結果, 計算式Hの値（小さい値のほうがよい）, 試行回数")
print("--------------------------------------------------------")
for r in result[:5]:
    df, subs = Auto_array(r[0]).get_dframe('q{}_{}')
    #print(df)
    print(f'H: {r[1]:.2f}, 試行回数: {r[2]}\n')
```

```python
import matplotlib.pyplot as plt
# 地図表示
# 色を割り当て red :00, blue: 01, yellow :10, green: 11
colors = list()
df, subs = Auto_array((result[0])[0]).get_dframe('q{}_{}')
for index, row in df.iterrows():
  #print('Index: ', index, 'q0: ', row['0'], 'q1: ', row['1'])
  color_num = row['0'] + 2*row['1']
  match color_num:
    case 0:
      colors.append('red')
    case 1:
      colors.append('blue')
    case 2:
      colors.append('yellow')
    case 3:
      colors.append('green')

#print(colors)

jp.pref_map(range(1,48), cols=colors, width=2.5)
```
